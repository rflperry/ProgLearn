```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gini(y1, y2):
    # Assuming binary y1, y2 elements
    p1 = np.mean(y1)
    p2 = np.mean(y2)
    return - (len(y1) * p1 * (1-p1) + len(y2) * p2 * (1-p2))

def grf_mse(y1, y2):
    p1 = np.mean(y1)
    p2 = np.mean(y2)
    return len(y1) * p1**2 + len(y2) * p2**2
```

```python
n = 20
y = np.random.binomial(1, 0.5, (n))

ginis = []
mses = []
for i in range(1, n-1):
    y1 = y[:i]
    y2 = y[i:]
    ginis.append(gini(y1, y2))
    mses.append(grf_mse(y1, y2))
    
plt.plot(list(range(1, n-1)), ginis, label='gini', ls='-', c='red')
plt.axvline(np.argmax(ginis)+1, label='gini max', ls='--', c='red')
plt.plot(list(range(1, n-1)), mses, label='mse', ls='-', c='blue')
plt.axvline(np.argmax(mses)+1, label='mse max', ls='--', c='blue')
plt.xticks(list(range(1, n-1)))
plt.legend()
plt.xlabel('split index')
plt.ylabel('Score to maximize')
plt.show()
```
